## **3.** ***Data Wrangling***

### Data Cleaning

````python
# Remove duplicates rows
booking_df.drop_duplicates(inplace = True)
````

````python
# Drop is_canceled column as we have reservation status with similar data
booking_df.drop('is_canceled', axis = 1, inplace = True)
````

````python
# Replace occurrences of the values 'Undefined' in the 'meal' column with 'No Meal
booking_df.replace({'meal': " 'Undefined', 'No Meal' "}, inplace = True)
````

````python
# Handle missing values of children column by filling the missing values with 0
booking_df.fillna({'children':0}, inplace = True)
````

````python
# Handle missing values of country column using mode
booking_df.fillna({'company': booking_df['country'].mode()[0]}, inplace = True)
````

````python
# Drop distribution channel column as it is similar to market segment
booking_df.drop('distribution_channel', axis =1, inplace = True)
````

````python
# Replace the value 0 and 1 in column 'is_repeated_guest' with 'No' and 'Yes'
booking_df.replace({'is_repeated_guest':{0:'No',1:'Yes'}}, inplace = True)
````

````python
# Handling missing values of column 'agent' by filling it with 'No agent'
booking_df.fillna({'agent': 'No agent'}, inplace = True)
````

````python
# Handling missing values of column 'company'
booking_df.fillna({'company':'No company'}, inplace = True)
````

````python
# Combine arrival year, month and day to single column(arrival_date)
arrival_date = booking_df['arrival_date_year'].astype(str) + '-' + booking_df['arrival_date_month'] + '-' + booking_df['arrival_date_day_of_month'].astype(str)
booking_df['arrival_date'] = pd.to_datetime(arrival_date)
````

````python
# Drop the arrival year, month and day columns as we created arrival_date column
booking_df.drop(['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month'], axis = 1, inplace = True)
````

````python
# Convert datatype of reservation status date to datetype
booking_df['reservation_status_date'] = pd.to_datetime(booking_df['reservation_status_date'])
````

````python
# Check for the outliers in column "lead_time"
# Create a box plot of the 'lead_time' feature
sns.boxplot(x = 'lead_time', data = booking_df)

# Show the plot
plt.show()
````

````python
# Dealing with outliers of lead time using IQR method
booking_df = booking_df[(booking_df.loc[:,'lead_time'] >= booking_df.loc[:,'lead_time'].quantile(0.05)) &
        (booking_df.loc[:,'lead_time'] <= booking_df.loc[:,'lead_time'].quantile(0.95))] #Removing the outlier
````

````python
# Handling outliers in market segment
booking_df = booking_df[booking_df['market_segment'] != 'Undefined']
````

````python
# Check for the outliers in column "adr"
# Create a box plot of the 'adr' feature
sns.boxplot(x = 'adr', data = booking_df)

# Show the plot
plt.show()
````

````python
# Dealing with outliers of adr using IQR method
booking_df = booking_df[(booking_df.loc[:,'adr'] >= booking_df.loc[:,'adr'].quantile(0.05)) &
        (booking_df.loc[:,'adr'] <= booking_df.loc[:,'adr'].quantile(0.95))]
````




###Data Aggregation


````python
# Aggregate the 'lead_time' column by calculating the mean lead time for each reserved room type and sorting to find room with the highest mean lead time
room_type_avg_lead_time = booking_df.groupby("reserved_room_type").agg({"lead_time":"mean"}).rename(columns={"lead_time":"avg_lead_time"}).sort_values(by = "avg_lead_time", ascending = False).reset_index()

# Print the aggregated dataframe to show the average lead time for each reserved room type
room_type_avg_lead_time
````

````python
#Finding top meal category using mode function
top_meal_category = booking_df['meal'].mode()[0]

# print the top category meal
print(f"Most of the peoples are prefering '{top_meal_category}' category meal.")
````

````python
# Finding booking count countrywise
booking_count_countrywise = booking_df['country'].value_counts()
print("Top 5 countries based on booking count as follows :")

# print first 5 rows to show the booking counts for each top 5 countries
booking_count_countrywise.head()
````

````python
# Find the cancellation count for each market segment where the reservation status is either 'Canceled' or 'No-Show'
market_segment_cancellation_count = booking_df[(booking_df['reservation_status'] == 'Canceled') |
                                (booking_df['reservation_status'] == 'No-Show')].groupby('market_segment').agg({'reservation_status':'count'}).rename(columns = {'reservation_status':'cancellation_count'}).sort_values(by = 'cancellation_count', ascending = False).reset_index()

# Print the resulting dataframe to view cancellation counts for each market segment
print(market_segment_cancellation_count)
````

````python
# Group the dataset by 'market_segment' and 'reserved_room_type', then count the occurrences of each combination. Also, sort the results and drop duplicates
top_room_type = booking_df.groupby(['market_segment', 'reserved_room_type']).size().reset_index(name='room_type_count').sort_values(by = ['market_segment', 'room_type_count'], ascending = [True, False]).drop_duplicates(subset=['market_segment'], keep='first').reset_index()

# Drop the 'index' column from the dataframe
top_room_type.drop('index', axis = 1, inplace = True)

# Display the dataframe with the top room type for each market segment with their counts
top_room_type
````

````python
# Group the data with cancellations by 'agent' column' and aggregate by cancellation count. Then, rename the column and sort the values by cancellation count
agent_cancellation = booking_df[booking_df['reservation_status'] != 'Check-Out'].groupby('agent').agg({'reservation_status':'count'}).rename(columns = {'reservation_status':'cancellation_count'}).sort_values(by = 'cancellation_count', ascending = False).reset_index()

# Remove rows where the 'agent' column has the value 'No agent'
agent_cancellation.drop(agent_cancellation[agent_cancellation['agent'] == 'No agent'].index, inplace = True)

# Filter the dataframe to keep only agents with more than 50 cancellations
agent_cancellation = agent_cancellation[agent_cancellation['cancellation_count']>50]

# Print the final dataframe with agents having more than 50 cancellations
agent_cancellation
````

````python
# Group the data with cancellations by 'deposit_type' column' and aggregate by cancellation count. Then, rename the column and sort the values by cancellation count
deposit_cancellation = booking_df[booking_df['reservation_status'] != 'Check-Out'].groupby('deposit_type').agg({'reservation_status':'count'}).rename(columns = {'reservation_status':'cancellation_count'}).sort_values(by = 'cancellation_count', ascending = False).reset_index()

# Print the resulting dataframe containing the cancellation count for each deposit type
deposit_cancellation
````

````python
# Group the required car parking data by 'reserved_room_type' and 'required_car_parking_spaces' and aggregate by count of 'required_car_parking_spaces'. Then remane the column and sort the values by 'required_car_parking_spaces' and count column
car_parking_room_type = booking_df[booking_df['required_car_parking_spaces'] != 0].groupby(['reserved_room_type','required_car_parking_spaces']).agg({'required_car_parking_spaces':'count'}).rename(columns = {'required_car_parking_spaces':'count'}).sort_values(by=['required_car_parking_spaces', 'count'], ascending=[True, False]).reset_index()

# Print the resulting dataframe with room type and requested parking spaces, along with the count
car_parking_room_type
````

````python
# Group the data with cancellations by 'customer_type' column' and aggregate by cancellation count. Then, rename the column and sort the values by cancellation count
customer_type_cancellation = booking_df[booking_df['reservation_status'] != 'Check-Out'].groupby('customer_type').agg({'reservation_status':'count'}).rename(columns = {'reservation_status':'cancellation_count'}).sort_values(by = 'cancellation_count', ascending = False).reset_index()

# Print the resulting dataframe containing the cancellation count for each customer type
customer_type_cancellation
````
